```python
# Creating dataframe from dictiionary
weather_data = {
    'day' : ['01/01/2023', '02/01/2023', '03/01/2023', '04/01/2023', '05/01/2023', '06/01/2023'],
    'temperature' : [32, 35, 28, 24, 32, 32],
    'windspeed': [6, 7, 2, 7, 4, 2],
    'event' : ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny']
}
df_dict = pd.DataFrame(weather_data)
df_dict
```

```python
# Creating dataframe using Tuple list.
# When creating using tuple, we need to provide column names
weather_data = [
    ('01/01/2023', 32, 6, 'Rain'),
    ('02/01/2023', 35, 2, 'Sunny'),
    ('03/01/2023', 28, 7, 'Snow')
]
df_tuple = pd.DataFrame(weather_data, columns = ['day', 'temperature', 'windspeed', 'event'])
df_tuple
```

```python
# Creating a dataframe using a list of dictionaries
weather_data = [
    {'day' : '01/01/2023', 'temperature' : 32, 'windspeed' : 6, 'event' : 'Rain' },
    {'day' : '02/01/2023', 'temperature' : 35, 'windspeed' : 2, 'event' : 'Sunny' },
    {'day' : '03/01/2023', 'temperature' : 28, 'windspeed' : 7, 'event' : 'Snow' },
]
df_li_dict = pd.DataFrame(weather_data)
df_li_dict
```
